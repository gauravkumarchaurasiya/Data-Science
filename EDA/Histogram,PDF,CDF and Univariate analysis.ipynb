```python
#we leart about 2d , 3d, 4d plots what about 1d plot(just only one feature)
#1d scatterplot of petal-length
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('iris_csv.csv')
iris
```

```python
iris_setosa=iris.loc[iris['class']=='Iris-setosa']
iris_setosa
```

```python
iris_virginica=iris.loc[iris['class']=='Iris-virginica']
iris_virginica
```

```python
print(iris_setosa['petallength'])
```

```python
plt.plot(iris_setosa["petallength"],np.zeros_like(iris_setosa['petallength']))
plt.plot(iris_virginica['petallength'],np.zeros_like(iris_virginica['petallength']))
# all overlapping each other
#disadvantage : very difficult to make sense as points are overlapping a lot
```

```python
# Histogram : tower like structure of 1d plot
```

```python
sns.FacetGrid(iris,hue="class",height=5)\
.map(sns.distplot,'petallength')\
.add_legend()
plt.show()
# conclusion setosa is well separted than other while there is overlapping over versicolor and virginica
```

```python
#CDF
    # tell about prob percentage of correctness 
    # eg : there are 82% of setosa flower that have pl<1.6
```

```python
counts, bin_edges=np.histogram(iris_setosa['petallength'],bins=10, density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)

#compute CDF
cdf=np.cumsum(pdf) # function to compute CDF in one line
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
