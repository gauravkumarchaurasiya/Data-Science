```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('iris_csv.csv')
```

```python
# Pair wise Scatter plot

# from ( sl,sw,pl,pw)  4C2 = 6

#     (sl,sw) (pl,pw)
#     (sl,pl) (pl,sw)
#     (sl,pw) (pw,sw)
```

```python
sns.set_style('whitegrid')
sns.pairplot(iris,hue='class',height=3)
plt.show()

# we can observer 4th dim using pair plot 
# obser: pl and pw we can separatable using logical if cond
```
