```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.title("Myplot")
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.grid()
```

```python
import numpy as np
t=np.arange(0,50,2)
plt.plot(t,"r^")
```

```python
t=np.arange(0,5,.2)
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.legend()
plt.show()
```

```python
plt.plot(t,t*t,linewidth=4,color='black')
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines=plt.plot(x1,y1,x2,y2)

plt.setp(lines[0],color='r',linewidth=3)
plt.setp(lines[1],'color','g','linewidth','2')
```

```python
def f(t):
    return np.exp(-t) * np.cos(2* np.pi * t)
t1=np.arange(0,5,.1)
t2=np.arange(0,5,.02)
plt.figure(1)
plt.subplot(211)
plt.plot(t1,f(t1),'b-')

plt.subplot(212)
plt.plot(t2,np.cos(2 * np.pi * t2),'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])

plt.subplot(212)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(211)
plt.title('Easy as 1 2 3')
```
