## Quantile-Quantile (Q-Q) plot

```python
# It answer question like X and Y are two distribution then Does X & Y have the same distribution?
# or it used to find does any distribution is normal or not by comparing with standard gaussian form
```

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
#N(0,1)
std_normal=np.random.normal(loc=0,scale=1,size=1000)
std_normal
```

```python
#0 to 100 the percentiles of std-normal
for i in range(0,100):
    print(i,np.percentile(std_normal,i))
```

```python
#generates 100 samples from N(20,5)
measurements= np.random.normal(loc=20,scale=5,size=100)
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()

# show X &Y are same distribution 
```

```python
# when we increase sample size 
measurements= np.random.normal(loc=20,scale=5,size=10000)
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
```

```python
#generates 100 sample from N(20,5)
measurements=np.random.uniform(low=-1,high=-10,size=100)
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()

# Observation : X & Y are NOT same distribution
```

```python
# On increase of sample size 
measurements=np.random.uniform(low=-1,high=-10,size=10000)
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
```
